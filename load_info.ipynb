```python
import pandas as pd
import re
```

```python
def read_txt(archivo): ## función para abrir y leer el archivo txt
    return open(archivo, encoding ='utf-8').readlines()
```

```python
texto=read_txt('./data/OFEI1204.txt')
```

```python
def clear_txt(txt): ## función para limpiar el texto 
    txt = [s.strip('\n') for s in txt] ## elimino los saltos de linea 
    txt = [s.replace(':',',').replace(" ",'') for s in txt] ## cambio ':' -> ',' & ' ' -> ''
    txt = list(filter(None, txt)) # elimino elementos vacios
    data = txt.pop(0) # elimino el primer elemento 
    print(txt)
    return txt ## returna una lista que contiene los datos
```

```python
texto_clear=clear_txt(texto)
```

```python
def convert_txt(data): ## función organiza la data para ser procesada como un dataframe tipo CSV
    indices = [] ## lista que almacena los indices que seran borrados
    for index,element in enumerate(data): ## recorremos la lista obtenemos el indice y elemento
        if re.match('^AGENTE,', element): ##  comparamos si el elemento inicia con la palabra Agente
            agente = (data[index]) ## si inicia obtenemos el elemento y lo guardamos
            indices.append(index) ## obtenemos el indice
        else:
            data[index] = agente + ',' + element ## si el elemento no inicia con la palabra agente lo adicionamos al string
    indices = sorted(indices, reverse = True)  ## Una vez finalizado el ciclo obtenemos los indices donde solamente
    for j in indices: ## estba palabra agente, borramos este elemento de la lista ya que no nos interesa
        del data[j] 
    return data   ## obtenemos los datos organizados por agente.planta,tipo,consumo
```

```python
texto_conv=convert_txt(texto_clear)
texto_conv
```

```python
def write_txt(datos,name): ## Función para escribir un archivo txt y guardarlo
    with open(name, 'w') as f:
        for item in datos:
            f.write("%s\n" % item)
```

```python
write_txt(texto_conv,'./data/OFEI1204_clear.txt') ## guardamos el archivo 
```
